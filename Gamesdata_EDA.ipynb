
##DISTPLOT

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import warnings
warnings.filterwarnings('ignore')
df=pd.read_excel('/content/games.xlsx')
plt.figure(figsize=(6,3.2),dpi=150)
plt.rc('xtick', labelsize=6)
plt.rc('ytick', labelsize=8)
sns.distplot(df['Accepted_Count'])
plt.show()
```

##HISTPLOT


```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import warnings
warnings.filterwarnings('ignore')
df=pd.read_excel('/content/games.xlsx')
plt.figure(figsize=(9,6))
sns.histplot(df['Year'],palette='Spectral',kde=True,binwidth=2)
plt.grid(axis='y',color='green',linestyle='--',linewidth=0.5)
plt.show()
```


##PIE CHART

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import warnings
warnings.filterwarnings('ignore')
df=pd.read_excel('/content/games.xlsx')
df.groupby(['Home_Team']).sum().plot( kind='pie', y='Home_Count',autopct='%1.0f%%')
plt.show()
```


##BARPLOT


##SCATTER PLOT

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import warnings
warnings.filterwarnings('ignore')
df=pd.read_excel('/content/games.xlsx')
df1 = df[df['Year']>2021]
x=df1['Home_Team']
y=df1['Home_Count']
plt.scatter(x, y,cmap='viridis')
plt.colorbar()
plt.show()
```


##PAIR PLOT


##CORRELATION HEATMAP

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import warnings
warnings.filterwarnings('ignore')
df=pd.read_excel('/content/games.xlsx')
df1 = df[df['Year']>2021]
sns.heatmap(df1.corr(),annot=True, cmap="YlGnBu")
plt.show()
```
